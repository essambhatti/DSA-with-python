```python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr
    mid = len(arr)//2
    left_side = arr[:mid]
    right_side = arr [mid:]

    left_sorted = merge_sort(left_side)
    right_sorted = merge_sort(right_side)

    return merge(left_sorted, right_sorted)

def merge(arr1,arr2):
    i=0
    j=0
    result = []
    while i < len(arr1) and j < len(arr2):
        if arr1[i] < arr2[j] :
            result.append(arr1[i])
            i+=1
        else:
            result.append(arr2[j])
            j+=1
    if i < len(arr1):
        for k in arr1[i:]:
            result.append(k)
    if j < len(arr2):
        for k in arr2[j:]:
            result.append(k)

    return result


print(merge_sort([11,3,6,2,1,4,5,8,7,10,9,12]))
```
