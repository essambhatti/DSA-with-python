```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[0]
    left_side = []
    right_side = []
    equal_side = []
    for i in arr:
        if i < pivot:
            left_side.append(i)
        elif i > pivot:
            right_side.append(i)
        else:
            equal_side.append(i)

    left_sorted = quick_sort(left_side)
    right_sorted = quick_sort(right_side)
    return left_sorted + equal_side + right_sorted
    

print(quick_sort([3,2,5,4,1]))
```
