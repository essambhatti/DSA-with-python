### 1929. Concatenation of Array

```python
class Solution(object):
    def getConcatenation(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        ans1 = [None]*len(nums)
        ans2 = [None]*len(nums)
        for i in range(len(ans1)):
            ans1[i] = nums[i]
            ans2[i] = nums[i]
        return ans1+ans2
```

### 2011. Final Value of variable after performing operations

```python
class Solution(object):
    def finalValueAfterOperations(self, operations):
        """
        :type operations: List[str]
        :rtype: int
        """
        X=0
        for i in operations:
            if "--" in i:
                X=X-1
            else:
                X=X+1
        return X
```

### 1431. Kids With the Greatest Number of Candies

```python
class Solution(object):
    def kidsWithCandies(self, candies, extraCandies):
        """
        :type candies: List[int]
        :type extraCandies: int
        :rtype: List[bool]
        """
        result=[False]*len(candies)
        for i in range(len(candies)):
            if candies[i] + extraCandies >= max(candies):
                result[i] = True
        return result
```

### 1389. Create Target Array in the Given Order

```python
class Solution(object):
    def createTargetArray(self, nums, index):
        """
        :type nums: List[int]
        :type index: List[int]
        :rtype: List[int]
        """
        target = []
        for i in range(len(nums)):
            target.insert(index[i] , nums[i])
        return target
```

### 136. Single Number

```python
class Solution(object):
    def singleNumber(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        output = 0
        for i in nums:
            output = output ^ i
        return output
```

### 27. Remove Element

```python
class Solution:
    def removeElement(self, nums: list[int], val: int) -> int:
        i=0
        while i < len(nums):
            if nums[i] == val:
                nums.remove(val)
            else:
                i+=1 
        return len(nums)
```

### 283. Move Zeroes

```python
class Solution:
    def moveZeroes(self, nums: list[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        i= 0
        while i < len(nums):
            number = nums[i]
            if number == 0:
                nums.remove(number)
                nums.append(number)
            i+=1
        return nums
```

### 66. Plus One

```python
class Solution:
    def plusOne(self, digits: list[int]) -> list[int]:
        string = ""
        for i in digits:
            string = string+str(i)
        number = int(string)
        number = number + 1
        final = str(number)
        result = []
        for i in final:
            result.append(int(i))
        return result
```

### 26. Remove Duplicates from Sorted Array

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        read = 1
        write = 1
        for i in range(len(nums)):
            if read > len(nums)-1:
                break
            elif nums[read] == nums[write-1]:
                read+=1
            else:
                nums[write] = nums[read]
                write+=1
                read+=1
        return write
```
