#### Task 1 : create a doubly linked list 

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

```

```python
class DoublyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, value):
        if not self.head:
            new_Node = Node(value)
            self.head = new_Node
            return
        new_Node = Node(value)
        new_Node.next = self.head
        self.head.prev = new_Node
        self.head = new_Node
    
    def insert_at_end(self, value):
        if not self.head:
            new_Node = Node(value)
            self.head = new_Node
            return
        new_Node = Node(value)
        curr = self.head 
        while curr.next != None:
            curr = curr.next
        curr.next = new_Node
        new_Node.prev = curr

    def insert_at_point(self, value_to_store, after_value):
        curr = self.head
        while curr and curr.data != after_value:
            curr = curr.next

        if not curr:
            print(f"Value {after_value} not found.")
            return

        new_Node = Node(value_to_store)
        new_Node.next = curr.next
        new_Node.prev = curr
        curr.next = new_Node
        if new_Node.next:  
            new_Node.next.prev = new_Node


    def traverse_forward(self):
        temp = self.head
        while temp:
            print(temp.data, end=" -> ")
            temp = temp.next
        print("None")

    def check_palindrome(self):
        if not self.head:
            print("Empty list â€” palindrome by default")
            return
        left = self.head
        right = self.head

        while right.next:
            right = right.next

        while left != right and left.prev != right:
            if left.data != right.data:
                print("This is not a palindrome")
                return
            left = left.next
            right = right.prev

        print("This is a palindrome")


    def reverse(self):
        if not self.head:
            print("List is empty")
            return

        curr = self.head
        prev_node = None

        while curr:
            prev_node = curr.prev
            curr.prev = curr.next
            curr.next = prev_node
            curr = curr.prev  # move to the next node (previous in original list)

        if prev_node:
            self.head = prev_node.prev
```

```python

ll = DoublyLinkedList()
ll.insert_at_end(3)
ll.insert_at_beginning(1)
ll.insert_at_beginning(2)
ll.insert_at_point(4,1)

ll.traverse_forward()
ll.reverse()
ll.traverse_forward()
```

```python
palindrome = DoublyLinkedList()
palindrome.insert_at_beginning(1)
palindrome.insert_at_beginning(2)
palindrome.insert_at_beginning(3)
palindrome.insert_at_beginning(2)
palindrome.insert_at_beginning(1)
palindrome.check_palindrome()

ll.check_palindrome()
```
