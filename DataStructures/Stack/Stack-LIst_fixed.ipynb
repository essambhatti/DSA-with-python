#### Fixed Sized Stack using python list

```python
class Stack:
    def __init__(self, size):
        self.stack = [None] * size
        self.top = -1
        self.size = size

    def is_empty(self):
        return self.top == -1
    
    def is_full(self):
        return self.top == self.size -1
    
    def push(self, value):
        if self.is_full() == True:
            return "Stack is Full"
        self.top +=1
        self.stack[self.top] = value
        return f"Pushed {value} in the start of stack"
    
    def pop(self):
        if self.is_empty():
            return "Stack is already empty"
        value = self.stack[self.top]
        self.stack[self.top] = None
        self.top -= 1
        return f"Popped {value} from the stack"
    
    def peek(self):
        return self.stack[self.top]
    
    def display(self):
        result = ""
        for i in self.stack:
            if i == None: 
                break
            result = result + str(i) + "->"

        return result[:-2]

```

```python
stack = Stack(3)

print(stack.push(1))
print(stack.push(2))
print(stack.push(3))
print(stack.push(3))

print(stack.peek())

print(stack.display())

print(stack.pop())
print(stack.pop())
print(stack.pop())
print(stack.pop())
```
