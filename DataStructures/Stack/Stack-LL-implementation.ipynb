### Stack implementation using Linked List

```python
class Node : 
    def __init__(self, value):
        self.data = value
        self.next = None
```

```python
class Stack : 
    def __init__(self):
        self.top = None

    def isempty(self):
        return self.top == None
    
    def push(self, value):
        new_Node = Node(value)
        new_Node.next = self.top
        self.top = new_Node
        
    def __str__(self):
        result = ""
        curr = self.top
        while curr != None:
            result = result + str(curr.data) + "->"
            curr = curr.next
        return result[:-2]
    
    def pop(self):
        if self.isempty():
            return "Stack Empty"
        previous_top = self.top
        self.top = self.top.next
        previous_top.next = None
        return previous_top.data

    def peek(self):
        if self.isempty():
            return "Stack Empty"
        return self.top.data

```

```python
stack = Stack()
stack.isempty()
```

```python
stack.push(2)
stack.push(3)
stack.push(4)
stack.isempty()
```

```python
print(stack)
```

```python
print(stack.pop())
print(stack)
```

```python
stack.peek()
```

#### Reversing a String Using Stack

```python
def reversed_string(string):
    stack = Stack()
    for i in string:
        stack.push(i)
    reverse_str = ""
    while stack.isempty() == False:
        char = stack.pop()
        reverse_str += str(char)
    return reverse_str
```

```python
reversed_string("Hello")
```

#### Performing a Pattern on a string using Stack

```python
def pattern_apply(string, pattern):
    string_stack = Stack()
    for i in string:
        string_stack.push(i)
    pattern_stack = Stack()  
    for i in pattern:
        if i == "u":
            curr = string_stack.pop()
            pattern_stack.push(curr)
        elif i == "r":
            curr = pattern_stack.pop()
            string_stack.push(curr)
    result = ""
    while string_stack.isempty() == False:
        result = string_stack.pop() + result
    return result
```

```python
pattern_apply("Hello","urr")
```
