```python
class Stack:
    def __init__(self):
        self.stack = []
        self.top = -1

    def is_empty(self):
        return len(self.stack) == 0
    def push(self, value):
        self.top+=1
        self.stack.append(value)
        return f"Pushed {value} in the stack"
    def pop(self):
        if self.is_empty():
            return "Stack is Already empty"
        self.stack.remove(self.stack[self.top])
        self.top-=1
    def peek(self):
        return self.stack[self.top]
    
    def display(self):
        result = ""
        for i in self.stack :
            result = result + str(i) + "->"
        # result = result[::-1]
        # return result[2:]

        return result[:-2]
```

```python
stack = Stack()

print(stack.push(1))
print(stack.push(2))
print(stack.push(3))
print(stack.push(3))

print(stack.peek())

print(stack.display())

print(stack.pop())
print(stack.pop())
print(stack.display())
print(stack.peek())
print(stack.pop())
print(stack.pop())
```
