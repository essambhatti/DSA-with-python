```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)

    def pop(self):
        if not self.stack:
            return None
        return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0
```

```python
def parenthese_check(string):
    pairs = {')': '(', ']': '[', '}': '{'}
    s = Stack()

    for ch in string:
        if ch in "([{":
            s.push(ch)
        elif ch in ")]}":
            if s.is_empty() or s.pop() != pairs[ch]:
                return False
    return s.is_empty()


print(parenthese_check("[]"))        
print(parenthese_check("[({})]"))    
print(parenthese_check("[({}]"))  
```
