```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, value):
        self.stack.append(value)

    def pop(self):
        if not self.stack:
            return None
        return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0

    def __str__(self):
        return str(self.stack)
```

```python
class UndoRedoSimulator:
    def __init__(self):
        self.undo_stack = Stack()
        self.redo_stack = Stack()

    def perform_action(self, action):
        self.undo_stack.push(action)
        self.redo_stack = Stack()
        print(f"Performed action {action}")

    def undo(self):
        if self.undo_stack.is_empty():
            return "Nothing to Undo"
        action = self.undo_stack.pop()
        self.redo_stack.push(action)
        print(f"Undo {action}")

    def redo(self):
        if self.redo_stack.is_empty():
            return "Nothing to redo"
        action = self.redo_stack.pop()
        self.undo_stack.push(action)
        print(f"Redo {action}")

    def show_state(self):
        print("Undo stack:", self.undo_stack)
        print("Redo stack:", self.redo_stack)
```

```python
sim = UndoRedoSimulator()
sim.perform_action("Type A")
sim.perform_action("Type B")
sim.undo()
sim.undo()
sim.redo()
sim.perform_action("Type C")
sim.show_state()
```
