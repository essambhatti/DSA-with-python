### Linked List class

* Head = None is the condition of empty linked list

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        self.n = 0

    def __len__(self):
        return self.n
    
    def insert_head(self, value):
        new_Node = Node(value)
        new_Node.next = self.head
        self.head = new_Node
        self.n+=1

    def __str__(self):
        curr = self.head
        result = ""
        while curr!=None:
            result += str(curr.value) + "->"
            curr = curr.next
        return result[:-2]
    
    def append(self,value):
        new_Node  = Node(value)
        if self.head == None:
            self.head = new_Node
        else:
            curr = self.head 
            #Logic 1:
            # while True:
            #     if curr.next == None:
            #         last_Node = curr
            #         break
            #     curr = curr.next
            # last_Node.next = new_Node

            #Logic 2:
            while curr.next !=None:
                curr = curr.next
            curr.next = new_Node

        #increment in length
        self.n+=1

    def insert_after(self,value, value_to_store):
        new_Node = Node(value_to_store)
        if self.head == None:
            return "Cannot perform this operation on empty linked list"
        curr = self.head
        while True:
            if curr == None:
                return "Cannot find the value"
            elif curr.value == value :
                new_Node.next = curr.next
                curr.next = new_Node
                self.n+=1
                break
            curr = curr.next

    def clear(self):
        self.head = None
        self.n=0

    def del_head(self):
        if self.head == None:
            return "Empty LL"
        self.head = self.head.next
        self.n-=1
    
    def pop(self):
        if self.head == None:
            return "Empty LL"
        elif self.head.next == None:
            self.del_head()
            return
        curr = self.head
        while curr.next.next != None:
            curr = curr.next
        curr.next = None
        self.n-=1

    def remove(self,value):
        curr = self.head
        if self.head == None:
            return "Empty LL"
        elif curr.value == value:
            self.del_head()
            return
        while curr.next.value != value:
            curr = curr.next
            if curr.next == None:
                return "Value not found"
        curr.next = curr.next.next
        self.n-=1

    def find(self,value):
        index = 0
        curr = self.head
        if self.head == None:
            return "Empty LL"
        while curr.value != value:
            if curr.next == None:
                return "Value not found"
            curr = curr.next
            index+=1
        return index
    def __getitem__(self,index):
        curr = self.head
        pos = 0
        while pos!=index:
            if curr.next == None:
                return "Index Error: index not found"
            curr = curr.next
            pos+=1

        return curr.value



        
```
