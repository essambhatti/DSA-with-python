```python
class Node : 
    def __init__(self, value):
        self.data = value
        self.next = None
```

```python
class Stack : 
    def __init__(self):
        self.top = None

    def isempty(self):
        return self.top == None
    
    def push(self, value):
        new_Node = Node(value)
        new_Node.next = self.top
        self.top = new_Node
        
    def __str__(self):
        result = ""
        curr = self.top
        while curr != None:
            result = result + str(curr.data) + "->"
            curr = curr.next
        return result[:-2]
    
    def pop(self):
        if self.isempty():
            return "Stack Empty"
        previous_top = self.top
        self.top = self.top.next
        previous_top.next = None

    def peek(self):
        if self.isempty():
            return "Stack Empty"
        return self.top.data

```

```python
stack = Stack()
stack.isempty()
```

```python
stack.push(2)
stack.push(3)
stack.push(4)
stack.isempty()
```

```python
print(stack)
```

```python
stack.pop()
print(stack)
```

```python
stack.peek()
```
